```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("..\Data\Raw\ACS\ACSDP052018.csv", skiprows=1)
df["year"] = 2018
df.head(5)
```

```python
list(df.columns)
```

```python
df = df[df.columns.drop(list(df.filter(regex='Margin of Error')))]
df.head(5)
```

```python
list(df.columns)
```

```python
df = df.rename(columns = {'Estimate!!SEX AND AGE!!Median age (years)' : 'median_age'})
```

```python
# total_pop_col = df[["Estimate!!SEX AND AGE!!Total population", "id"]]
# df = pd.merge(df, total_pop_col, how="left", on="id")
# df.head(5)
```

```python
df.columns = df.columns.str.replace('Estimate!!SEX AND AGE!!Total population_y', 'total_pop')
df.head(5)
```

```python
list(df.columns)
```

```python
df = df[["year", "id", "Estimate!!SEX AND AGE!!Total population",  'Percent Estimate!!SEX AND AGE!!Total population!!Female', "Percent Estimate!!RACE!!Total population!!One race!!White", 'Percent Estimate!!RACE!!Total population!!One race!!Black or African American', 'Percent Estimate!!RACE!!Total population!!One race!!American Indian and Alaska Native',  'Percent Estimate!!RACE!!Total population!!One race!!Asian', 'Percent Estimate!!RACE!!Total population!!One race!!Native Hawaiian and Other Pacific Islander', 'Percent Estimate!!RACE!!Total population!!Two or more races',  'Percent Estimate!!HISPANIC OR LATINO AND RACE!!Total population!!Hispanic or Latino (of any race)']]
df.head()
```

```python
df['fips'] = df['id'].str[-5:]
df.head(5)
```

```python
df = df.drop(columns = ["id"])
df.head()
```

```python
df = df.rename(columns={'Estimate!!SEX AND AGE!!Total population':'total_pop', 'Percent Estimate!!SEX AND AGE!!Total population!!Female':'pct_female', 'Percent Estimate!!RACE!!Total population!!One race!!White':'pct_white', "Percent Estimate!!RACE!!Total population!!One race!!Black or African American":"pct_black", "Percent Estimate!!RACE!!Total population!!One race!!American Indian and Alaska Native":"pct_native_american", "Percent Estimate!!RACE!!Total population!!One race!!Asian":"pct_asian", "Percent Estimate!!RACE!!Total population!!One race!!Native Hawaiian and Other Pacific Islander":"pct_native_hawaiian_pacific_islander", "Percent Estimate!!RACE!!Total population!!Two or more races":"pct_two_or_more_races", "Percent Estimate!!HISPANIC OR LATINO AND RACE!!Total population!!Hispanic or Latino (of any race)":"pct_hispanic"})
df.head()
```

```python
df['pct_female'] = df['pct_female'].apply(lambda x: x*0.01)
df['pct_white'] = df['pct_white'].apply(lambda x: x*0.01)
df['pct_black'] = df['pct_black'].apply(lambda x: x*0.01)
df['pct_native_american'] = df['pct_native_american'].apply(lambda x: x*0.01)
df['pct_asian'] = df['pct_asian'].apply(lambda x: x*0.01)
df['pct_native_hawaiian_pacific_islander'] = df['pct_native_hawaiian_pacific_islander'].apply(lambda x: x*0.01)
df['pct_two_or_more_races'] = df['pct_two_or_more_races'].apply(lambda x: x*0.01)
df['pct_hispanic'] = df['pct_hispanic'].apply(lambda x: x*0.01)

df.head()
```

```python
df.to_csv("..\Data\Clean\ACS\ACSDP052018_clean.csv")
```
