```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Data/Raw/nonACSDemographics/censusIntercensalCountyPopRaceSexEthnicity2000_2010.csv", encoding = "ISO-8859-1")

df.head()
```

```python
df.shape
```

```python
df["fips"] = df.STATE * 1000 + df.COUNTY
df = df.drop(columns=["SUMLEV", "STATE", "COUNTY", "STNAME", "CTYNAME", "ESTIMATESBASE2000", "CENSUS2010POP"])
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
df["group"] = ""
```

```python
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 0),'group'] = "total_pop"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 1),'group'] = "white"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 2),'group'] = "black"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 3),'group'] = "native_american"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 4),'group'] = "asian"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 5),'group'] = "native_hawaiian_pacific_islander"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 0) & (df["RACE"] == 6),'group'] = "two_or_more_races"
df.loc[(df["SEX"] == 0) & (df["ORIGIN"] == 2) & (df["RACE"] == 0),'group'] = "hispanic"
df.loc[(df["SEX"] == 2) & (df["ORIGIN"] == 0) & (df["RACE"] == 0),'group'] = "female"

df = df[df.group != ""]
```

```python
df = df.drop(columns=["ORIGIN", "SEX", "RACE"])
df.head(10)
```

```python
df = df.melt(id_vars=["fips", "group"], 
                               var_name="year",
                               value_name="POPESTIMATE")
df.year = df.year.str.replace("POPESTIMATE","")
df = df.rename(columns = {"POPESTIMATE":"total_pop"})
df.head()
```

```python
df = df[df.fips % 1000 != 0]
df = df[df.fips > 0]
df["year"] = df["year"].astype(int)
df = df[df.year <= 2009]
df.head(10)
```

```python
df = df.pivot_table(index=['year','fips'], columns=['group']).reset_index()
df.columns = df.columns.map('_'.join).str.strip('_')
df.head()
```

```python
df = df.rename(columns = {'total_pop_total_pop' : 'population'})
df.columns = df.columns.str.replace('total_pop_', '')
df.head()
```

```python
df["pct_asian"] = df.asian / df.population
df["pct_black"] = df.black / df.population
df["pct_native_american"] = df["native_american"] / df.population
df["pct_native_hawaiian_pacific_islander"] = df["native_hawaiian_pacific_islander"] / df.population
df["pct_two_or_more_races"] = df["two_or_more_races"] / df.population
df["pct_white"] = df.white / df.population
df["pct_hispanic"] = df.hispanic / df.population
df["pct_female"] = df.female / df.population

df.head()
```

```python
df = df.drop(columns=["asian", "black", "native_american", "native_hawaiian_pacific_islander", "two_or_more_races", "white", "population", "female", "hispanic"])
```
