```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(io = "../Data/Raw/nonACSDemographics/SAIPECountyPovertyAndIncome2009.xls", skiprows = 4, sheet_name="Unemployment Med HH Income")

df.head()
```

```python
list(df.columns)
```

```python
df = df[["State FIPS", "County FIPS", "Name", "Poverty Percent All Ages", "Median Household Income"]]
df.head()
```

```python
df = df[df["County FIPS"] > 0]
df.head()
```

```python
df["State FIPS"] = pd.to_numeric(df["State FIPS"])
df["year"] = 2009
df["fips"] = df["State FIPS"]*1000 + df["County FIPS"]
df = df.drop(columns = ["State FIPS", "County FIPS", "Name"])
df = df.rename(columns = {"Median Household Income":"per_capita_income_2009"})
df = df.rename(columns = {"Poverty Percent All Ages":"poverty_rate"})
df.head()
```

```python
df.to_csv("../Data/Clean/nonACSDemographics/SAIPE2009_clean.csv")
```

```python
df = pd.read_fwf('../Data/Raw/nonACSDemographics/SAIPECountyPovertyAndIncome2002.txt', header = None)
df.head()
```

```python
list(df.columns)
```

#### Documentation here: https://www2.census.gov/programs-surveys/saipe/technical-documentation/file-layouts/state-county/2002-estimate-layout.txt

```python
df = df[[0, 1, 5, 20, 29]]
df.head()
```

```python
df = df[df[1] > 0]
df.head()

df[0] = pd.to_numeric(df[0])
df["fips"] = df[0]*1000 + df[1]
df["year"] = 2002
df = df.drop(columns = [0, 1, 29])
df = df.rename(columns = {20:"per_capita_income_2002"})
df = df.rename(columns = {5:"poverty_rate"})
df.head()
```

```python
df.to_csv("../Data/Clean/nonACSDemographics/SAIPE2002_clean.csv")
```
