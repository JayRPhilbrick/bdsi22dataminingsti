```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(io = "../Data/Raw/nonACSDemographics/SAIPECountyPovertyAndIncome2009.xls", skiprows = 2, sheet_name="est09ALL")

df.head()
```

```python
list(df.columns)
```

```python
df = df[["State FIPS", "County FIPS", "Name", "Poverty Percent All Ages", "Median Household Income"]]
df.head()
```

```python
df = df[df["County FIPS"] > 0]
df.head()
```

```python
df["State FIPS"] = pd.to_numeric(df["State FIPS"])
df["fips"] = df["State FIPS"]*1000 + df["County FIPS"]
df = df.drop(columns = ["State FIPS", "County FIPS", "Name"])
df.head()

df.to_csv("../Data/Clean/nonACSDemographics/SAIPE2009_clean.csv")
```
