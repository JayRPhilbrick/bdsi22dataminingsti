```python
df_1 = pd.read_csv("../Data/Raw/CDC/CDCSTDoverall2000-2020.csv",skiprows = 4)
```

```python
df_1.head()
```

```python
df_1.columns = df_1.columns.str.lower()
df_1.columns = df_1.columns.str.replace(' ','_')
```

```python
df_1.loc[df_1["year"].str.startswith('2020',na=False), "year"] = "2020"
```

```python
data_to_be_removed = ["cases","rate_per_100000"]
```

```python
for col in data_to_be_removed:
    df_1.loc[df_1[col].str.contains("Data",na = False), col] = np.nan
    df_1[col] = df_1[col].str.replace(',','')
df_1[['cases','year','rate_per_100000']] = df_1[['cases','year','rate_per_100000']].apply(pd.to_numeric)
```

```python
searchfor = ['Gonorrhea', 'Chlamydia']
df_1_refine = df_1[df_1.indicator.str.contains('|'.join(searchfor))]
```

```python
df_1 = df_1_refine.pivot_table(index=['year','fips'], columns=['indicator']).reset_index()
# df.columns = df.columns.map('_'.join).str.strip('_')
df_1.head()
```
