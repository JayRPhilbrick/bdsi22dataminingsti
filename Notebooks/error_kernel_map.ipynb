```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('../Data/Final/pred_vals_errors.csv', index_col = 0)
cols = ['gonorrhea','chlamydia','early_non-primary_non-secondary_syphilis',
        'early_primary_and_secondary_syphilis']
```

```python
for i in cols[0:2]:
    plt.figure(figsize = (15,8))
    g = sns.kdeplot(data = d,x = "adj_errors_"+i)
    plt.title("Kernel density plot of " + i)
    plt.xlim(-0.025, 0.03)
    plt.savefig("../Output/PosterPresentation/Errors/kde_" + i + '.png',dpi = 300)
```
