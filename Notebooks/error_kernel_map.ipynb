```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
    counties = json.load(response)
import plotly.express as px
import plotly
```

```python
d = pd.read_csv('../Data/Final/model1_pred_vals_errors.csv', index_col = 0)
cols = ['gonorrhea','chlamydia','early_non-primary_non-secondary_syphilis',
        'early_primary_and_secondary_syphilis']
```

```python
for i in cols[0:2]:
    plt.figure(figsize = (12,8))
    g = sns.kdeplot(data = d,x = "adj_errors_"+i)
    plt.title("Kernel density plot of " + i + ' errors (2020)')
    plt.xlim(-0.004, 0.004)
    plt.xlabel('adjusted errors ' + i.replace('_',' '))
    plt.axvline(np.mean(d['adj_errors_'+i]), color='purple', label = 'mean adjusted error')
    plt.axvline(0, color='black', linestyle = '--')
    plt.legend()
    plt.savefig("../Output/PosterPresentation/Errors/kde_" + i + '.png',dpi = 300)
```
