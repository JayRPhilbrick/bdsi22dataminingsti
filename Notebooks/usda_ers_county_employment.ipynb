```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(io = "../Data/Raw/nonACSDemographics/USDAERSCountyUnemployment2000_2021.xlsx", skiprows = 4, sheet_name = "Unemployment Med HH Income")
df.head()
```

```python
list(df.columns)
```

```python
df = df.drop(columns = ["Area_name", "State", 'Rural_urban_continuum_code_2013', 'Urban_influence_code_2013', 'Metro_2013', 'Median_Household_Income_2020', 'Med_HH_Income_Percent_of_State_Total_2020'])
df_all = df
list(df.columns)
```

```python
df = df[df.columns.drop(list(df.filter(regex='^Employed')))]
df = df[df.columns.drop(list(df.filter(regex='^Civilian_labor_force')))]
df = df[df.columns.drop(list(df.filter(regex='^Unemployed')))]
df.head()
```

```python
list(df.columns)
```

```python
df = df.melt(id_vars=["FIPS_code"], 
                              var_name="year", 
                              value_name="Unemployment_rate")
df.year = df.year.str.replace("Unemployment_rate_","")
df_unemployment_rate = df
df.head()
```

```python
df_all.head()
df_employment = df_all[df_all.columns.drop(list(df_all.filter(regex='^Civilian_labor_force')))]
df_employment = df_employment[df_employment.columns.drop(list(df_employment.filter(regex='^Unemployed')))]
df_employment = df_employment[df_employment.columns.drop(list(df_employment.filter(regex='^Unemployment')))]
df_employment.head()

df_employment = df_employment.melt(id_vars=["FIPS_code"], 
                              var_name="year", 
                              value_name="Employed")
df_employment.year = df_employment.year.str.replace("Employed_","")
df_employment.head()
```

```python
df = pd.merge(df_employment, df_unemployment_rate, how = "inner", on=["FIPS_code", "year"])
df = df[df.FIPS_code > 0]
df = df[df.FIPS_code % 1000 != 0]
df = df.rename(columns = {"FIPS_code":"fips"})
df = df.rename(columns = {"Unemployment_rate":"unemployment_rate"})
df = df.rename(columns = {"Employed":"employed_persons"})
df.head()
```

```python
df["year"] = df["year"].astype(int)
df = df[df.fips <= 2009]
df.dtypes
```

```python
df.to_csv("../Data/Clean/nonACSDemographics/usdaERSCountyEmployment_clean.csv")
```
