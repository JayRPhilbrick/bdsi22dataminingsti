```python
import pandas as pd
import numpy as np
import os
from functools import reduce
from tabulate import tabulate
```

```python
total = []
folder = '../Data/Clean'
print(os.getcwd())
for dirs in os.listdir(folder):
    if not dirs.endswith(('.txt', 'checkpoints', 'HPS', 'ACS', 'USAFacts', 'graphics', 'ility')):
        #print(dirs)
        for files in os.listdir(os.path.join(folder,dirs)):
            if not files.endswith(('.txt', 'checkpoints', 'STD2019_2020_clean.csv')):
                k = pd.read_csv(os.path.join(folder,(os.path.join(dirs,files))), 
                                sep = ',', index_col = 0)
                k = k[(k['year'] >= 2002) & (k['year'] <= 2020)].reset_index(drop = True)
                k['fips'] = k['fips'].astype(str).str.zfill(5)
                total.append(k)
                print(os.path.join(folder, os.path.join(dirs, files)))
                print(k.shape)
                print(k['year'].value_counts())
    
folder2 = '../Data/Intermediate'
for files2 in os.listdir(folder2):
    if not (files2.endswith(('.txt', 'checkpoints', 'hps_processed.csv'))):
        l = pd.read_csv(os.path.join(folder2,files2), sep = ',', index_col = 0)
        l = l[(l['year'] >= 2002) & (l['year'] <= 2020)].reset_index(drop = True)
        l['fips'] = l['fips'].astype(str).str.zfill(5)
        total.append(l)
        print(files2)
        print(l.shape)
        print(l['year'].value_counts())
```

```python
for index, i in enumerate(total):
    if 'year' in i.columns:
        #i['year'] = i['year'].astypes('int64')
        print(index, ':    ', i['year'].dtypes)
        
        if 'geography' in i.columns:
            i.drop(columns = 'geography', inplace = True)
            print('Done')
        if 'county_name' in i.columns:
            i.drop(columns = 'county_name' , inplace = True)
            print('done')
        #print(indexi[['year', 'fips']].head(10))
```

```python
cov = total[0].columns.intersection(total[1].columns)
df = pd.merge(total[0], total[1], on = cov.to_list(), how = 'outer')
yrs = range(2017, 2020)
for i in yrs:
    lsttot = []
    lst2 = df[df['year'] == i]['fips'].unique()
    lsttot.append(lst2)

tc1 = total[1]['fips'].unique()
tc0 = total[0]['fips'].unique()
tct = []
tct.append(tc1)
tct.append(tc0)
tct = [x for xs in tct for x in xs]
tct = list(set(tct))

print(total[1]['year'].value_counts())
print(total[0]['year'].value_counts())
print(df['year'].value_counts())
      
print(len(tct))
lsttot = [x for xs in tct for x in xs]
print(list(set(lsttot) - set(tct)))
```
