```python
import pandas as pd
import numpy as np
import os
from functools import reduce
```

```python
total = []
folder = '../Data/Clean'

for dirs in os.listdir(folder):
    if not dirs.endswith(('.txt', 'checkpoints', 'HPS', 'ACS', 'USAFacts')):
        #print(dirs)
        for files in os.listdir(os.path.join(folder,dirs)):
            if not files.endswith(('.txt', 'checkpoints', 'STD2019_2020_clean.csv')):
                total.append(pd.DataFrame(pd.read_csv(os.path.join(folder,(os.path.join(dirs,files))), sep = ',', index_col = 0)))
                print(os.path.join(folder, os.path.join(dirs, files)))
    
folder2 = '../Data/Intermediate'
for files2 in os.listdir(folder2):
    if not (files2.endswith(('.txt', 'checkpoints', 'hps_processed.csv'))):
            total.append(pd.DataFrame(pd.read_csv(os.path.join(folder2,files2), sep = ',', index_col = 0)))
            print(files2)
            print(pd.DataFrame(pd.read_csv(os.path.join(folder2,files2), sep = ',', index_col = 0)).shape)
```

```python
for i in total:
    print(i.shape)
    if 'geography' in i.columns:
        i.drop(columns = 'geography', inplace = True)
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left, right, on=['year','fips'],
                                            how='outer'), total)
print(df_merged.shape)
#list(df_merged.columns)
```

```python
rates = [1.25, 1.23, 1.23, 1.22, 1.19, 1.16, 1.14, 1.14]
df_merged['per_capita_income22'] = None
years = list(range(2013, 2021))

for index, i in enumerate(years): 
    df_merged.loc[df_merged['year'] == i, 'per_capita_income22'] = df_merged.loc[df_merged['year'] == i, 'per_capita_income_' + str(i)] * rates[index]
    df_merged.drop(columns = 'per_capita_income_' + str(i), inplace = True)
    #print(df_merged.loc[df_merged['year'] == i, 'per_capita_income22'])
#print(df_merged.loc[:,'per_capita_income_2018'].dtypes)
```

```python
print(df_merged.shape)
list(df_merged.columns)
```

```python
df_merged['state_fips'] = (np.floor(df_merged['fips'] / 1000)).astype('int64')
print(df_merged['state_fips'].unique())
df_merged[df_merged['state_fips']==11].head()
```

```python
hps = pd.DataFrame(pd.read_csv('../Data/Intermediate/hps_processed.csv', sep = ',', index_col = 0))
```

```python
df_tot = pd.merge(df_merged, hps, on = ['state_fips', 'year'], how = 'outer')
```
