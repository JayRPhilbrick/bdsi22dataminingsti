```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Data/Raw/CDC/CDCSTDoverall2000-2020.csv",skiprows=4)
```

```python
df.head()
```

```python
pd.unique(df['Year'])
```

```python
df.columns= df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('/','_')
df.head()
```

```python
df.loc[df["year"].str.startswith('2020',na=False), "year"] = "2020"
```

```python
df.head()
```

```python
data_to_be_removed = ["cases", "rate_per_100000"]
```

```python
for col in data_to_be_removed:
    df.loc[df[col].str.contains("Data",na = False), col] = np.nan
    df[col] = df[col].str.replace(',','')
```

```python
df[["year","fips","cases","rate_per_100000"]] = df[["year","fips","cases","rate_per_100000"]].apply(pd.to_numeric)
```

```python
df = df.pivot_table(index=['year','geography','fips'], columns=['indicator']).reset_index()
df.columns = df.columns.map('_'.join).str.strip('_')
df.head()
```

```python
df.columns= df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace('/','_')
df.columns = df.columns.str.replace(',','')
df.head()
```

```python
df_new = pd.read_csv("../Data/clean/CDC/disagRaceSexSTD2019_2020_clean.csv")
```

```python
df_merge = pd.merge(df,df_new,how = 'outer',on = ['year','fips','geography'])
```
