```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("..\Data\Raw\ACS\ACSDP052020.csv", skiprows=1)
df["year"] = 2020
df.head(5)
```

```python
list(df.columns)
```

```python
df = df[df.columns.drop(list(df.filter(regex='Margin of Error')))]
df.head(5)
```

```python
list(df.columns)
```

```python
df = df.rename(columns = {'Estimate!!SEX AND AGE!!Total population!!Median age (years)' : 'median_age'})
```

```python
total_pop_col = df[["Estimate!!SEX AND AGE!!Total population", "id"]]
df = df[df.columns.drop(list(df.filter(regex='Estimate')))]
# df = df[df.columns.drop(list(df.filter(regex='Percent!!SEX AND AGE!!Total population')))]
df = df[df.columns.drop(list(df.filter(regex='RACE!!Two or more races!!')))]
df = df[df.columns.drop(list(df.filter(regex='RACE!!One race!!')))]
df = pd.merge(df, total_pop_col, how="left", on="id")
list(df.columns)
```

```python
df.columns = df.columns.str.replace('Estimate!!SEX AND AGE!!Total population', 'total_pop')
df.columns = df.columns.str.replace('Percent!!', 'pct_')
df.columns = df.columns.str.replace('SEX AND AGE!!', 'sex_age_')
df.columns = df.columns.str.replace('HISPANIC OR LATINO AND RACE!!', 'ethnicity_')
df.columns = df.columns.str.replace('RACE!!', 'race_')
df.columns = df.columns.str.replace('Male', 'male')
df.columns = df.columns.str.replace('Female', 'female')
df.columns = df.columns.str.replace('CITIZEN, VOTING AGE POPULATION!!Citizen, 18 and over population!!', 'voting_age_citizen_')
df.columns = df.columns.str.replace('Race alone or in combination with one or more other races', 'one_or_more_race')
list(df.columns)
```

```python
# df = df.drop(columns=["pct_sex_age_Median age (years)"])
df.columns = df.columns.str.replace('Under 5 years', 'under_5')
df.columns = df.columns.str.replace('5 to 9 years', '5_to_9')
df.columns = df.columns.str.replace('10 to 14 years', '10_to_14')
df.columns = df.columns.str.replace('15 to 19 years', '15_to_19')
df.columns = df.columns.str.replace('20 to 24 years', '20_to_24')
df.columns = df.columns.str.replace('25 to 34 years', '25_to_34')
df.columns = df.columns.str.replace('35 to 44 years', '35_to_44')
df.columns = df.columns.str.replace('45 to 54 years', '45_to_54')
df.columns = df.columns.str.replace('55 to 59 years', '55_to_59')
df.columns = df.columns.str.replace('60 to 64 years', '60_to_64')
df.columns = df.columns.str.replace('65 to 74 years', '65_to_74')
df.columns = df.columns.str.replace('75 to 84 years', '75_to_84')
df.columns = df.columns.str.replace(' years and over', '+')
df.columns = df.columns.str.replace('!!male', '_male')
df.columns = df.columns.str.replace('!!female', '_female')


list(df.columns)
```

```python
# df = df.drop(columns = ["pct_sex_age_65+.1"])
# df = df.drop(columns = ["pct_sex_age_18+.1"])
df = df.drop(columns=["pct_race_Total population!!One race"])
df = df.drop(columns=["pct_race_Total population!!Two or more races"])
# df = df.drop(columns=["pct_race_One race"])
df = df.drop(columns=["pct_race_Total population"])
# df = df.drop(columns=["pct_race_Two or more races"])
df = df.drop(columns=["pct_ethnicity_Total population"])
df.columns = df.columns.str.replace('!!Total population', '')
df.columns = df.columns.str.replace('!!Total population', '')
df.columns = df.columns.str.replace('Total population', '')
df.columns = df.columns.str.replace('!!Hispanic or Latino (of any race)!!', 'hispanic_')
df.columns = df.columns.str.replace('!!Not Hispanic or Latino!!', 'not_hispanic_')


list(df.columns)
```

```python
df = df[df.columns.drop(list(df.filter(regex='^pct_ethnicity_not_hispanic')))]
df = df[df.columns.drop(list(df.filter(regex='^pct_ethnicity_!!Hispanic or Latino (of any race)!!')))]
df = df[df.columns.drop(list(df.filter(regex='^pct_voting_age')))]

df = df.drop(columns=["pct_ethnicity_!!Hispanic or Latino (of any race)!!Mexican"])
df = df.drop(columns=["pct_ethnicity_!!Hispanic or Latino (of any race)!!Puerto Rican"])
df = df.drop(columns=["pct_ethnicity_!!Hispanic or Latino (of any race)!!Cuban"])
df = df.drop(columns=["pct_ethnicity_!!Hispanic or Latino (of any race)!!Other Hispanic or Latino"])
df = df.drop(columns=["pct_ethnicity_!!Not Hispanic or Latino"])
df = df.drop(columns=["pct_Total housing units"])
df = df.drop(columns=["pct_CITIZEN, VOTING AGE POPULATION!!Citizen, 18 and over population"])

df.columns = df.columns.str.replace('White', 'white')
df.columns = df.columns.str.replace('Black or African American', 'black')

df.columns = df.columns.str.replace('American Indian and Alaska Native', 'native_american')
df.columns = df.columns.str.replace('Asian', 'asian')
df.columns = df.columns.str.replace('Native Hawaiian and Other Pacific Islander', 'native_hawaiian_pacific_islander')
df.columns = df.columns.str.replace('Some other race', 'other_race')

df

list(df.columns)
```

```python
df = df.drop(columns = ["pct_sex_age__male",  'pct_sex_age__female', "pct_sex_age_", 'pct_sex_age_!!Sex ratio (males per 100 females)'])

df.columns = df.columns.str.replace('pct_sex_age_!!', 'pct_sex_age_')

df = df[df.columns.drop(list(df.filter(regex='.1$')))]
df = df[df.columns.drop(list(df.filter(regex='^pct_race_!!One race!!')))]
df = df[df.columns.drop(list(df.filter(regex='^pct_race_!!Two or more races!!')))]


df = df.drop(columns = ["pct_sex_age_Median age (years)", "pct_sex_age_Under 18 years", "pct_sex_age_18+!!Sex ratio (males per 100 females)", "pct_sex_age_65+!!Sex ratio (males per 100 females)"])

df.columns = df.columns.str.replace('pct_one_or_more_race!!', 'pct_one_or_more_race_')

df = df.rename(columns = {'pct_ethnicity_!!Hispanic or Latino (of any race)' : 'pct_ethnicity_hispanic'})
```
