```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../Data/Raw/CDC/HIV/CDCHIVcountyage2017_2020.csv",skiprows=11)
print(data.shape)
data.head()
```

```python
data = data.rename(columns = {'Cases':'cases',
                              'Rate per 100000':'ratePer100000',
                             'Indicator':'indicator',
                             'Year':'year',
                             'Geography':'geography',
                             'FIPS':'fips',
                             'Percent':'percent',
                             'Age Group':'ageGroup'})
data.head(30)
```

```python
data.loc[data["year"].str.startswith('2020'), "year"] = "2020"
```

```python
data_to_be_removed = ["cases", "ratePer100000", "percent"]
```

```python
for col in data_to_be_removed:
    data.loc[data[col].str.contains("Data",na = False), col] = np.nan
    data[col] = data[col].str.replace(',','')
```

```python
data[["year","fips","cases","ratePer100000","percent"]] = data[["year","fips","cases","ratePer100000","percent"]].apply(pd.to_numeric)
```

```python
data.dtypes
# data.head()
```

```python
data = data.pivot_table(index=['year','geography','fips'], columns=['ageGroup', 'indicator']).reset_index()
data.columns = data.columns.map('_'.join).str.strip('_')
data.head()
```

```python
data.to_csv('../Data/clean/CDC/HIVcountyage_clean.csv')
```
