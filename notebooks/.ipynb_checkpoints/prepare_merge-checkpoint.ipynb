```python
import numpy as np
import pandas as pd
import os
```

```python
df1 = pd.read_csv('../Data/Intermediate/COVIDcounts_county_processed.csv', index_col = 0)
print(df1.shape)
df1.head()
```

```python
labs20, labs21, labs22= [], [], []

for column in df1.columns[4:]:
    if column.startswith('2020'):
        labs20.append(column)
    if column.startswith('2021'):
        labs21.append(column)
    if column.startswith('2022'):
        labs22.append(column)
```

```python
df1 = pd.DataFrame(np.repeat(df1.values, 3, axis = 0), columns = df1.columns)
df1.head()
```

```python
df1['year'] = None
for index in df1.index:
    if index % 3 == 0:
        df1.iloc[index, -1] = 2020
    elif index % 3 == 1:
        df1.iloc[index, -1] = 2021
    else:
        df1.iloc[index, -1] = 2022
df1.head()
```

```python
df1['cases'] = None
for index in df1.index:
    if index % 3 == 0:
        df1.iloc[index, -1] = df1.loc[index,labs20].sum(axis = 0)
    elif index % 3 == 1:
        df1.iloc[index, -1] = df1.loc[index,labs21].sum(axis = 0)
    else:
        df1.iloc[index, -1] = df1.loc[index,labs22].sum(axis = 0)
```

```python
df = df1[['year', 'fips', 'county_name', 'state', 'statefips', 'cases']]
print(df.dtypes)
df[['year','fips','statefips','cases']] = df[['year','fips','statefips','cases']].apply(pd.to_numeric)
print(df.dtypes)
df
```

```python
df.to_csv('../Data/Clean/COVIDcounty_cases_clean.csv')
```
