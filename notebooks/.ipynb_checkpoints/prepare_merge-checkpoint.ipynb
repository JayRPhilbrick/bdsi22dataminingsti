```python
import numpy as np
import pandas as pd
import os
```

```python
df1 = pd.read_csv('../Data/clean/COVIDcounts_county_clean.csv', index_col = 0)
df1.head()
```

```python
labs20, labs21, labs22= [], [], []
for column in df1.columns[4:]:
    if column.startswith('2020'):
        labs20.append(column)
    if column.startswith('2021'):
        labs21.append(column)
    if column.startswith('2022'):
        labs22.append(column)
```

```python
for index, column in enumerate(df1):
    df1[index,'year'] = df1[labs20].sum(axis = 1)
    
df1.head()
```
