```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
### Updated
```

### Counties rate per 1000000 Chlamydia

```python
tot_df = []
for j,i in enumerate(np.arange(2019, 2021, 1)):
    tot_df.append(pd.read_csv('../Data/Raw/CDC/CDCSTDcountysexrace'+ 
                              str(i) +'.csv', sep = ',', skiprows = 4))
df = pd.concat(tot_df, axis = 0)
```

```python
df.loc[df["Year"].str.startswith('2020',na=False), "Year"] = "2020"
```

```python
data_to_be_removed = ["Cases","Rate per 100000"]
```

```python
for col in data_to_be_removed:
    df.loc[df[col].str.contains("Data",na = False), col] = np.nan
    df[col] = df[col].str.replace(',','')
df[['Cases','Year','Rate per 100000']] = df[['Cases','Year','Rate per 100000']].apply(pd.to_numeric)
```

```python
df.head()
```

```python
searchfor = ['Chlamydia']
df_refine = df[df.Indicator.str.contains('|'.join(searchfor))]
```

```python
df_race1 = df_refine.groupby(['Year','Geography','FIPS'])['Rate per 100000','Cases'].mean().reset_index()
df_race1.head()
```

```python
df = df_race1.pivot_table(index=['Geography','FIPS'], columns=['Year']).reset_index()
# df.columns = df.columns.map('_'.join).str.strip('_')
df.head()
```

```python
df['diff_rate'] = df['Rate per 100000'][2020] -df['Rate per 100000'][2019]
```

```python
df.head()
```

```python
plt.figure(figsize = (15,8))
g = sns.kdeplot(data = df,x = "diff_rate")
plt.title("Chlamydia")
plt.savefig("../Output/EDA/Chlamydia_rate_diff_kernel.png",dpi = 300)
# plt.axhline(y=0,c="black")
plt.show()
```

```python
df.sort_values("diff_rate")[:10]
```

```python
df.sort_values("diff_rate",ascending= False)[:10]
```

### Scatterplot of Chlamydia

```python
data_pop = pd.read_csv("../Data/Intermediate/ACSDP05_merge.csv")
```

```python
data_pop.head()
```

```python
data_pop = data_pop[['year','fips','total_pop']]
```

```python
data_pop = data_pop.loc[data_pop.year.isin([2019,2020]),]
```

```python
data_pop
```

```python
df_race1.columns = df_race1.columns.str.lower()
df_race1
```

```python
df1 = pd.merge(df_race1,data_pop,on = ['year','fips'])
```

```python
df1['adj_cases'] = df1['cases']/df1['total_pop']
```

```python
df1
```

```python
df_ref1 = df1.pivot_table(index=['fips'], columns=['year']).reset_index()
df_ref1
```

```python
df_ref1['adj_cases'][2019]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
data_temp1 = pd.DataFrame((df_ref1['rate per 100000'][2019],df_ref1['rate per 100000'][2020])).transpose()
data_temp1["color"]="high ratio between 2020 and 2019"
data_temp1["color"].iloc[data_temp1[2019]>=np.tan(3*math.pi/8)*data_temp1[2020]] = "low ratio between 2020 and 2019"
data_temp1["color"].iloc[(data_temp1[2020]<=data_temp1[2019]) & (data_temp1[2019] <np.tan(3*math.pi/8)*data_temp1[2020])] = "relative low ratio between 2020 and 2019"
data_temp1["color"].iloc[(np.tan(math.pi/8)*data_temp1[2020]<=data_temp1[2019]) & (data_temp1[2019] <data_temp1[2020])] = "relative high ratio between 2020 and 2019"
#data_temp["color"].iloc[(np.tan(math.pi/8)*data_temp[2020]>data_temp[2019])] = "d"

data_temp1
```

```python
plt.figure(figsize = (15,8))
g = sns.scatterplot(x = df_ref1['rate per 100000'][2019],y = df_ref1['rate per 100000'][2020],size = df_ref1.total_pop[2019],hue = data_temp1["color"],sizes=(10, 1000),alpha=0.5)
xpoints = ypoints = plt.xlim()

x1=0
y1=0
#col = np.where(rate20>np.tan(math.pi *3/8)*rate19, "r", "m")
s1 = np.tan(math.pi *3/8)
plt.axline((x1,y1), slope = s1)
plt.grid()


x=np.linspace(0,1000,1000)
y=s1*x

plt.plot(xpoints, ypoints, linestyle='--', color='k', lw=3, scalex=False, scaley=False)
plt.plot(x, y, linestyle='--', color='k', lw=3, scalex=False, scaley=False)


s2 = np.tan(math.pi /8)
plt.axline((x1,y1), slope = s2, linestyle='--', color='k', lw=3)


#plt.plot(x,y)

#plt.scatter(x, y, c =col)
#plt.scatter(x, y>np.tan(math.pi *3/8)*x,color= 'red', label='strong')
#plt.scatter(x1, x1>y1>np.tan(math.pi /8)*x1,color= 'green', label='weak')

# sns.lineplot(x = [1,2],y =[1,2])
plt.title("Chlamydia rate per 100000")
plt.savefig("../Output/EDA/Chlamydia_scatterplot4.png",dpi = 300)
# plt.axhline(y=0,c="black")
plt.show()

```

```python
plt.figure(figsize = (15,8))
g = sns.scatterplot(x = df_ref1['rate per 100000'][2019],y = df_ref1['rate per 100000'][2020],size = df_ref1.total_pop[2019],sizes=(10, 1000),alpha=0.5)
xpoints = ypoints = plt.xlim()
plt.plot(xpoints, ypoints, linestyle='--', color='k', lw=3, scalex=False, scaley=False)
# sns.lineplot(x = [1,2],y =[1,2])
plt.title("Chlamydia")
plt.savefig("../Output/EDA/Chlamydia_scatterplot.png",dpi = 300)
# plt.axhline(y=0,c="black")
plt.show()
```
